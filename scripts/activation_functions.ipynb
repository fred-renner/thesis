```python
import numpy as np
import matplotlib.pyplot as plt


# ReLU function
def relu(x):
    return np.maximum(0, x)


# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


# Generate x values
x = np.linspace(-10, 10, 400)

# Calculate y values for ReLU and Sigmoid functions
y_relu = relu(x)
y_sigmoid = sigmoid(x)

# Plot ReLU
plt.figure(figsize=(3,2.5))
plt.plot(x, y_relu, label="ReLU")
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig("/Users/fred/phd/thesis/images/relu_activation_function.pdf")

# Plot Sigmoid
plt.figure(figsize=(3,2.5))
plt.plot(x, y_sigmoid, label="Sigmoid", color="orange")
plt.title("Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("/Users/fred/phd/thesis/images/sigmoid_activation_function.pdf")

print("Plots saved as PDF files.")
```
