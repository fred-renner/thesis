```python
import numpy as np
import matplotlib.pyplot as plt


# ReLU function
def relu(x):
    return np.maximum(0, x)


# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


# Generate x values
x = np.linspace(-10, 10, 400)

# Calculate y values for ReLU and Sigmoid functions
y_relu = relu(x)
y_sigmoid = sigmoid(x)

# Plot ReLU
plt.figure(figsize=(3,2.5))
plt.plot(x, y_relu, label="ReLU")
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.savefig("/Users/fred/phd/thesis/images/relu_activation_function.pdf")

# Plot Sigmoid
plt.figure(figsize=(3,2.5))
plt.plot(x, y_sigmoid, label="Sigmoid", color="orange")
plt.title("Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("/Users/fred/phd/thesis/images/sigmoid_activation_function.pdf")

print("Plots saved as PDF files.")
```

```python
import numpy as np
from scipy.optimize import fsolve

# Sigmoid function with slope adjustment
def sigmoid(x, slope=1):
    return 1 / (1 + np.exp(-slope * x))

# Equation to solve: sigmoid(x, slope=50) - 0.95 = 0
def equation_to_solve(x):
    return sigmoid(x, slope=50) - 0.95

# Use fsolve to find the root of the equation
x_value = fsolve(equation_to_solve, 0)[0]

print(f"The x-value where y=0.95 for the sigmoid function with a slope of 50 is: {x_value}")

# Plot Sigmoid with slope 50 and mark the point where y=0.95
x = np.linspace(-1, 1, 400)
y_sigmoid_slope_50 = sigmoid(x, slope=50)

plt.figure(figsize=(4,3))
plt.plot(x, y_sigmoid_slope_50, label="Sigmoid with slope 50", color="orange")
plt.axhline(y=0.95, color='red', linestyle='--', label="y=0.95")
plt.axvline(x=x_value, color='blue', linestyle='--', label=f"x={x_value:.3f}")
plt.scatter([x_value], [0.95], color='green')
# plt.title("Sigmoid Activation Function with Slope 50")
plt.xlabel("x")
plt.ylabel("Sigmoid")
plt.legend()
plt.grid(True)
plt.savefig('/Users/fred/phd/thesis/images/sigmoid_slope_50.pdf')

print("Plot saved as PDF file.")
def unscale_value(value,):
    value -=             -0.0011581625588125781
    value /= 1.3231924376614288e-07
    return value
print(unscale_value(x_value))
```
