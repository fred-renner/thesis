```python
from pdf2image import convert_from_path
import pathlib
import matplotlib.pyplot as plt
import numpy as np
import glob

# https://stackoverflow.com/questions/53255432/saving-a-grid-of-heterogenous-images-in-python


def savegrid(ims, rows=None, cols=None, fill=False, showax=False, suffix=0):
    if rows is None != cols is None:
        raise ValueError("Set either both rows and cols or neither.")

    if rows is None:
        rows = len(ims)
        cols = 1

    gridspec_kw = {"wspace": -0.6, "hspace": 0} if fill else {}
    fig, axarr = plt.subplots(rows, cols, dpi=500, gridspec_kw=gridspec_kw)

    if fill:
        bleed = 0
        fig.subplots_adjust(
            left=bleed, bottom=bleed, right=(1 - bleed), top=(1 - bleed)
        )

    for ax, im in zip(axarr.ravel(), ims):
        ax.imshow(im)
        if not showax:
            ax.set_axis_off()

    kwargs = {"bbox_inches":'tight',"pad_inches": 0.0} 

    fig.savefig(f"/Users/fred/phd/thesis/images/tomats_inputs_{suffix}.png", **kwargs)


# files = []
# for f in glob.iglob("/Users/fred/Downloads/tomatos_inputs/*.pdf"):
#     files += [f]

# # move legend to end
# files.append(
#     files.pop(
#         files.index("/Users/fred/Downloads/tomatos_inputs/tomatos_input_legend.pdf")
#     )
# )
# for f in files:
#     print(f)
# ok lets just do it by hand

first = [
    "/Users/fred/phd/thesis/images/tomatos_inputs/pt_h1_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/phd/thesis/images/tomatos_inputs/pt_h2_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/phd/thesis/images/tomatos_inputs/with_legend/pt_hh_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/phd/thesis/images/tomatos_inputs/eta_h1_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/phd/thesis/images/tomatos_inputs/eta_h2_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/phd/thesis/images/tomatos_inputs/eta_hh_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/phd/thesis/images/tomatos_inputs/phi_h1_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/phd/thesis/images/tomatos_inputs/phi_h2_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/phd/thesis/images/tomatos_inputs/phi_hh_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/phd/thesis/images/tomatos_inputs/m_h1_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/phd/thesis/images/tomatos_inputs/m_h2_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/phd/thesis/images/tomatos_inputs/m_hh_NOSYS.SR_xbb_2.pdf",
]


ims = [np.array(convert_from_path(f, 500)[0]) for f in first]
savegrid(ims, rows=4, cols=3, suffix=0, fill=True)


second = [
    "/Users/fred/Downloads/tomatos_inputs/pt_j1_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/Downloads/tomatos_inputs/pt_j2_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/Downloads/tomatos_inputs/blank.pdf",
    "/Users/fred/Downloads/tomatos_inputs/eta_j1_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/Downloads/tomatos_inputs/eta_j2_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/Downloads/tomatos_inputs/lead_xbb_score_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/Downloads/tomatos_inputs/phi_j1_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/Downloads/tomatos_inputs/phi_j2_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/Downloads/tomatos_inputs/tomatos_input_legend.pdf",
    "/Users/fred/Downloads/tomatos_inputs/E_j1_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/Downloads/tomatos_inputs/E_j2_NOSYS.SR_xbb_2.pdf",
    "/Users/fred/Downloads/tomatos_inputs/blank.pdf",
]

ims = [np.array(convert_from_path(f, 500)[0]) for f in second]
savegrid(ims, rows=4, cols=3, suffix=1, fill=True)
```
